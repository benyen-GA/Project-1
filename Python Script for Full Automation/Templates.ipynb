## Updating Smartsheet from OCR / UCR <a id="ss-ucr"></a>

```python
#Generic Python package for Updating any Smartsheet
#Need to remember to add in column (Last Update Date) in any new Smartsheet or else code will not run!

import os
import shutil
from datetime import datetime
import smartsheet
import pandas as pd

def update_ss(df, sheet_id, column_mapping):
    access_token = "c49V20P7AmHDbeZAfJ1kq3UutLEyeb8Vvyk5s"
    smartsheet_client = smartsheet.Smartsheet(access_token)
    sheet = smartsheet_client.Sheets.get_sheet(sheet_id)
    
    column_map = {column.title: column.id for column in sheet.columns}
    
    if 'MFG_LOC_CODE' not in column_map:
        raise ValueError("Column 'MFG_LOC_CODE' not found in Smartsheet.")

    batch_size = 300
    updated_rows = [] 
    updated_row_ids = set()
    missing_ids = []
    assessment_ids = []

    print("----------------------------------------------------------")
    print(f"Script is now updating {column_mapping}!")
    
    for index, row in df.iterrows():
        name_value = row['MFG_LOC_CODE']
        
        row_id = None
        for smartsheet_row in sheet.rows:
            mfg_loc_code_cell = next((cell for cell in smartsheet_row.cells if cell.column_id == column_map['MFG_LOC_CODE']), None)
            if mfg_loc_code_cell and mfg_loc_code_cell.value == name_value:
                row_id = smartsheet_row.id
                break
                
        if row_id is not None and row_id not in updated_row_ids:
            updated_row_ids.add(row_id)
            
            updated_row = smartsheet.models.Row()
            updated_row.id = row_id
            
            for smartsheet_col, df_col in column_mapping.items():
                cell = smartsheet_client.models.Cell()
                cell.column_id = column_map[smartsheet_col]
                cell.value = row[df_col]    
                cell.strict = False
                updated_row.cells.append(cell)                
                updated_rows.append(updated_row)
                
            if 'Last Update Date' in column_map:
                update_date = smartsheet_client.models.Cell()
                update_date.column_id = column_map['Last Update Date']
                update_date.value = datetime.today().strftime('%d-%b-%Y')
                update_date.strict = False
                updated_row.cells.append(update_date)
            else:
                print('Last Update Date column not found!')

            if len(updated_rows) == batch_size:
                smartsheet_client.Sheets.update_rows(sheet_id, updated_rows)
                updated_rows = []

        else:
            missing_ids.append(row['MFG_LOC_CODE'])
            assessment_ids.append(row['Assessment ID'])

    if updated_rows:
        smartsheet_client.Sheets.update_rows(sheet_id, updated_rows)
        print(f"Final Batch of rows updated in Column {column_mapping}")
    else:
        print(f"No more rows to update")

    today_date = datetime.now().strftime("%d %b %Y")
    tracker_filename = f'Update Report (Tracker) - {today_date}.txt'
    with open(tracker_filename, "w") as file:
        if missing_ids:
            file.write('Errors found:\n')
            file.write('Missing ID\tMissing Assessment\n')
            unique_missing_ids = set(missing_ids)
            for i, missing_id in enumerate(unique_missing_ids):
                file.write(f'{missing_id}\t{assessment_ids[missing_ids.index(missing_id)]}\n')
        else:      
            file.write('No errors\n')

        num_rows_df = len(df)
        num_rows_updated_ss = len(updated_row_ids)

        if num_rows_df == num_rows_updated_ss:
            completeness_result = f"Completeness check: All rows in DataFrame have been successfully updated in Smartsheet. {num_rows_df} is updated!"
        else:
            completeness_result = f"Completeness check failed: Number of rows in DataFrame ({num_rows_df}) does not match the number of rows updated in Smartsheet ({num_rows_updated_ss})."

        file.write(completeness_result + '\n')

    print(f"Script has finished updating {column_mapping}!")
    print("----------------------------------------------------------")

    # Move the file to a desired location
    new_location = "/Users/byen1/Library/CloudStorage/Box-Box/ARO - WORKPLAN-INDUSTRY TOOL/FFC - Smartsheet Update"
    new_filepath = os.path.join(new_location, tracker_filename)
    shutil.move(tracker_filename, new_filepath)
```

#### Function for pulling out sheet ID

```python
import smartsheet 
import pandas as pd

def sheet_data(sheet_id):
    access_token = 'c49V20P7AmHDbeZAfJ1kq3UutLEyeb8Vvyk5s'
    
    smartsheet_client = smartsheet.Smartsheet(access_token)
    sheet = smartsheet_client.Sheets.get_sheet(sheet_id)
    
    columns = sheet.columns
    rows = sheet.rows
    
    data = []
    
    column_titles = [column.title for column in columns]
    data.append(column_titles)
    
    for row in rows:
        row_data = [cell.value for cell in row.cells]
        data.append(row_data)
        
    df = pd.DataFrame(data[1:],columns = data[0])
    df = df.rename(columns = {i:title for i, title in enumerate(data[0])})
    
    return df 
```

## Sending of email announcement (Bronze / Red-TRV) <a id="send-rating"></a>

```python
def send_email_announcement(df,ass_type,ass_year,attach_type,rating_col,report_n,
                            vrf_date,next_vrf_d,af_date):
    
    import time
    from pathlib import Path
    from appscript import app, k
    from mactypes import Alias
    
    outlook = app('Microsoft Outlook')

    for index, row in df.iterrows():
        code = row['MFG_LOC_CODE']
        factory = row['FACTORY NAME']
        report_name = row[report_n]

        rating = row[rating_col]
        VRF_date = row[vrf_date]#SLCP FY24 VRF Date
        next_VRF_date = row[next_vrf_d]#SLCP FY25 Due Date
        overall_rating = row['OVERALL RATING']
        slcp_rating = row['SLCP Overall']
        fem_rating = row['FEM Overall']
        zdhc_rating = row['ZDHC Overall']
        fla_rating = row['FLA Overall']
        fir_rating = row['FIR Labor H&S Overall']
        ua_rating = row['UA Labor Overall']
        bw_rating = row['BW Non SLCP Overall']
        ncat_rating = row['NCAT Labor H&S Overall']
        effective_date = row[af_date]

        factory_contact_1 = row['Factory Contact 1 (RSC Primary contact in group)']
        factory_contact_2 = row['Factory Contact 2-Primary Contact in Fty Location']
        factory_contact_3 = row['Factory Contact 3-Factory RSC Team']
        factory_contact_4 = row['Factory Contact 4-Factory RSC Team']

        #Licensee
        licensee_1 = row['Licensee Contact']
        licensee_2 = row['Licensee Contact 2'] 
        licensee_3 = row['Licensee Contact 3']
        licensee_4 = row['Licensee Contact 4']
        licensee_5 = row['Licensee Contact 5']
        licensee_6 = row['Licensee Contact 6']
        licensee_7 = row['Licensee Contact 7']
        licensee_8 = row['Licensee Contact 8']
        licensee_9 = row['Licensee Contact 9']
        licensee_10 = row['Licensee Contact 10']

        #Factory GM
        gm_1 = row['Factory-GM/Leadership 1']
        gm_2 = row['Factory-GM/Leadership 2']
        gm_3 = row['Factory-GM/Leadership 3']

        regional = row['RSC-Partner Responsibility Director']
        risk = row['RSC-RISK & ISSUES MANAGER']

        rsc_pm = row['RSC-RISK & ISSUES CONSULTANT 1']
        rsc_pm_2 = row['RSC-RISK & ISSUES CONSULTANT 2']

        rsc_lead = row['RSC-Partner Lead']

        internal_contact_1 = row['Nike Internal Contact (PM/MSM Team) 1']
        internal_contact_2 = row['Nike Internal Contact (PM/MSM Team) 2']
        internal_contact_3 = row['Nike Internal Contact (PM/MSM Team) 3']
        
        if rating == "Bronze":
            subject = f'NIKE Rating Announcement - {ass_year} - {code}/{factory}'
            body = bronze(factory, code, VRF_date, next_VRF_date, overall_rating, slcp_rating,
                          fem_rating, zdhc_rating, fla_rating, fir_rating, ua_rating,
                          bw_rating, ncat_rating, effective_date,ass_type)
            
        elif rating == "Red":
            subject = f'NIKE Rating Announcement and TRV Enrollment - {ass_year} - {code}/{factory}'
            body = red_trv(factory, code, VRF_date, next_VRF_date, overall_rating, slcp_rating,
                           fem_rating, zdhc_rating, fla_rating, fir_rating, ua_rating,
                           bw_rating, ncat_rating, effective_date, ass_type)
        
        elif rating == "Red" & attach_type == "zdhc":
            subject = f'NIKE Rating Announcement - {ass_year} - {code} / {factory}'
            body = zdhc_red(factory, code, VRF_date, next_VRF_date, overall_rating, slcp_rating,
                           fem_rating, zdhc_rating, fla_rating, fir_rating, ua_rating,
                           bw_rating, ncat_rating, effective_date, ass_type)
        
        else:
            raise ValueError("Invalid Rating")
            
        to_recip = get_recipients(factory_contact_1, factory_contact_2, factory_contact_3,
                                  factory_contact_4,
                                  licensee_1, licensee_2, licensee_3,
                                  licensee_4, licensee_5, licensee_6, licensee_7, licensee_8, licensee_9,
                                  licensee_10, gm_1, gm_2, gm_3, regional, risk,
                                  rsc_pm, rsc_pm_2, rsc_lead,
                                  internal_contact_1, internal_contact_2, internal_contact_3, rating)
        
        # Filter out empty email addresses
        to_recip = [recipient for recipient in to_recip if recipient]

        # Create new email message
        msg = outlook.make(new=k.outgoing_message, with_properties={k.subject: subject, k.content: body})
        # Add recipients
        for recipient in to_recip:
            msg.make(new=k.to_recipient, with_properties={k.email_address: {k.address: recipient}})
            
        bcc_recipient = 'compliancesupport@nike.com'           
        msg.make(new=k.bcc_recipient, with_properties={k.email_address:{k.address:bcc_recipient}})
        
        if attach_type == "zdhc":
            attachment_path = get_attachment_path_zdhc(rating, report_name)
        elif attach_type == "slcp":
            attachment_path = get_attachment_path(rating, report_name)
        else:
            attachment_path = get_attachment_path_fem(rating, report_name)
        
        if attachment_path is not None: 
            attachment_paths = attachment_path.split(',')
            for path in attachment_paths:
                attachment = Alias(str(Path(path.strip())))
                msg.make(new=k.attachment, with_properties={k.file:attachment})   
                
        # Open and activate the email message
        msg.activate()
        msg.open()

        # Pause before sending next email
        time.sleep(8)
```
